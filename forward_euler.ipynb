```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 0
dt = 0.01
T = 10.0
z0 = np.array([[1,0]])
Z = z0
A = 1.0; B = 0.5; C=5;
```

```python
while t<T:
    f = np.array([z0[0,1],(1/A)*(-B*z0[0,1]-C*z0[0,0])])
    z1 = z0 + dt * f
    z0 = z1
    Z = np.append(Z,z1,axis=0)
    t += dt
```

```python
plt.plot(Z[:,0],lw=2)
plt.show()
```
